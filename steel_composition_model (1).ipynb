```python
import os
import cv2
import numpy as np
import xml.etree.ElementTree as ET
from collections import defaultdict
from albumentations import (
    HorizontalFlip,
    VerticalFlip,
    RandomRotate90,
    RandomBrightnessContrast,
    RandomScale,
    ShiftScaleRotate,
    Compose,
)
from imblearn.over_sampling import RandomOverSampler

# Path to the directory containing images and annotations
data_dir = "/content/drive/MyDrive/steel composition.v2i.voc/train"  # Update this path

# Initialize a dictionary to store class counts and image file names
class_counts = defaultdict(list)

# Read images and annotations
for xml_file in os.listdir(data_dir):
    if xml_file.endswith(".xml"):
        # Load the XML file
        tree = ET.parse(os.path.join(data_dir, xml_file))
        root = tree.getroot()

        # Extract image information
        filename = root.find("filename").text
        img_path = os.path.join(data_dir, filename)

        # Store class names
        for obj in root.findall("object"):
            class_name = obj.find("name").text
            class_counts[class_name].append(img_path)

# Prepare data for oversampling
X = []
y = []

# Collect all images and their corresponding labels
for class_name, images in class_counts.items():
    X.extend(images)  # Append image paths
    y.extend([class_name] * len(images))  # Append class names

# Perform oversampling
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(np.array(X).reshape(-1, 1), np.array(y))

# Define augmentations
augmentation = Compose([
    HorizontalFlip(p=0.5),
    VerticalFlip(p=0.5),
    RandomRotate90(p=0.5),
    RandomBrightnessContrast(p=0.5),
    RandomScale(scale_limit=0.2, p=0.5),
    ShiftScaleRotate(p=0.5)
])

# Augment and save new images
output_dir = "augmented_data"  # Directory to save augmented images
os.makedirs(output_dir, exist_ok=True)

for img_path, label in zip(X_resampled.flatten(), y_resampled):
    img = cv2.imread(img_path)
    if img is not None:
        # Apply augmentations
        augmented = augmentation(image=img)
        augmented_img = augmented['image']

        # Save the augmented image
        base_name = os.path.basename(img_path)
        save_path = os.path.join(output_dir, f"augmented_{base_name.split('.')[0]}_{label}.jpg")
        cv2.imwrite(save_path, augmented_img)

print("Data augmentation and oversampling completed.")
```

```python
import os
import torch
import torchvision.transforms as transforms
from torchvision import datasets, models
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Directory containing augmented data
data_dir = "/content/augmented_data"

# Data transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize images to match the model input
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # Normalize with ImageNet statistics
])

# Load the dataset
dataset = datasets.ImageFolder(root=data_dir, transform=transform)
train_loader = DataLoader(dataset, batch_size=16, shuffle=True)

# Load a pre-trained model
model = models.resnet18(pretrained=True)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, len(dataset.classes))  # Adjust the final layer for your classes
model = model.to(device)

# Loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 20
for epoch in range(num_epochs):
    model.train()  # Set model to training mode
    running_loss = 0.0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # Print the average loss for this epoch
    print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {running_loss / len(train_loader):.4f}")

# Save the model
torch.save(model.state_dict(), 'steel_composition_model.pth')

print("Training completed and model saved.")
```
